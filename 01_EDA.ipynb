```python
import pandas as pd       # Importa a biblioteca pandas, usada para leitura, análise e manipulação de dados
import numpy as np        # Importa a biblioteca NumPy, útil para cálculos numéricos e manipulação de arrays
import matplotlib.pyplot as plt  # Importa o módulo pyplot do Matplotlib, usado para criar gráficos e visualizações
import seaborn as sns     # Importa o Seaborn, biblioteca baseada no Matplotlib para gráficos estatísticos mais bonitos e informativos

# Carregar o dataset a partir de um arquivo CSV e armazenar em um DataFrame chamado 'df'
df = pd.read_csv("ecommerce_sales.csv")

# Exibir as primeiras 5 linhas do dataset (por padrão)
# Isso ajuda a ter uma visão geral da estrutura dos dados — colunas, tipos e valores iniciais
df.head()
```

```python
# Exibe informações gerais sobre o DataFrame
df.info()
# Gera estatísticas descritivas de todas as colunas e as transpõe para facilitar a leitura
df.describe(include='all').T
```

```python
df.isna().sum().sort_values(ascending=False)
df.duplicated().sum()
```

```python
import seaborn as sns               # Importa o Seaborn, usado para criar gráficos estatísticos com boa aparência
import matplotlib.pyplot as plt     # Importa o módulo pyplot do Matplotlib, usado para configurar e exibir os gráficos

# Seleciona apenas as colunas numéricas do DataFrame (tipos int64 e float64)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Cria uma paleta de cores com base no número de colunas numéricas selecionadas
colors = sns.color_palette("viridis", len(num_cols))

# Loop para gerar um histograma para cada coluna numérica
for i, col in enumerate(num_cols):
    plt.figure(figsize=(6, 4))  # Define o tamanho da figura (6x4 polegadas)
    
    # Cria o histograma da coluna 'col' com linha de densidade (kde=True)
    # Define a cor, número de bins (30) e borda branca nas barras
    sns.histplot(df[col], kde=True, color=colors[i], bins=30, edgecolor='white')
    
    # Define o título do gráfico, formatando o nome da coluna (substitui '_' por espaço e coloca em formato título)
    plt.title(f"Distribuição de {col.replace('_', ' ').title()}", fontsize=12)
    
    # Define o rótulo do eixo X com o nome da variável
    plt.xlabel(col.replace('_', ' ').title())
    
    # Define o rótulo do eixo Y
    plt.ylabel("Frequência")
    
    # Adiciona uma grade (linhas de apoio) no eixo Y com estilo tracejado e transparência leve
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    
    # Ajusta automaticamente o layout para evitar sobreposição de elementos
    plt.tight_layout()
    
    # Exibe o gráfico na tela
    plt.show()

```

```python
# Loop sobre uma lista de colunas numéricas específicas
for col in ['marketing_spend', 'website_traffic', 'avg_price', 'monthly_sales']:
    
    # Cria uma figura com 1 linha e 2 colunas de gráficos (lado a lado)
    # figsize define o tamanho total da figura (10 de largura x 3 de altura)
    fig, axes = plt.subplots(1, 2, figsize=(10, 3))
    
    # --- Gráfico 1: Violin Plot ---
    # Mostra a distribuição dos dados combinando boxplot e densidade
    # 'inner="quartile"' desenha as linhas dos quartis dentro do violino
    sns.violinplot(x=df[col], ax=axes[0], color='lightgreen', inner='quartile')
    
    # Define o título do primeiro gráfico
    axes[0].set_title(f"Violin Plot - {col.replace('_', ' ').title()}")
    
    # --- Gráfico 2: Curva de Densidade (KDE Plot) ---
    # Mostra a estimativa da densidade de probabilidade da variável
    sns.kdeplot(df[col], ax=axes[1], shade=True, color='teal', linewidth=2)
    
    # Define o título do segundo gráfico
    axes[1].set_title(f"Densidade - {col.replace('_', ' ').title()}")
    
    # Ajusta automaticamente os espaçamentos para evitar sobreposição de elementos
    plt.tight_layout()
    
    # Exibe os dois gráficos para a coluna atual
    plt.show()

```

```python
import numpy as np  # Importa o NumPy, usado para operações numéricas e manipulação de arrays

# Calcula a matriz de correlação apenas para as colunas numéricas do DataFrame
corr = df.corr(numeric_only=True)

# Cria uma máscara para ocultar a metade superior da matriz (já que a correlação é simétrica)
mask = np.triu(np.ones_like(corr, dtype=bool))

# Define o tamanho da figura do gráfico
plt.figure(figsize=(9, 7))

# Cria o heatmap (mapa de calor) da matriz de correlação
sns.heatmap(
    corr,                # Dados de correlação
    mask=mask,           # Máscara para exibir apenas a parte inferior da matriz
    annot=True,          # Mostra os valores numéricos dentro de cada célula
    cmap="coolwarm",     # Define o esquema de cores (azul → negativo, vermelho → positivo)
    fmt=".2f",           # Formata os valores com duas casas decimais
    linewidths=0.5,      # Define espessura das linhas divisórias entre as células
    cbar_kws={"shrink": 0.8, "label": "Correlação"}  # Configura a barra de cores
)

# Define o título do gráfico
plt.title("Mapa de Correlação (Triangular Superior)", fontsize=14, fontweight='bold')

# Ajusta rotação dos rótulos do eixo X e Y para melhor visualização
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Ajusta automaticamente o layout para evitar sobreposição
plt.tight_layout()

# Exibe o gráfico na tela
plt.show()
```

```python
import seaborn as sns               # Importa o Seaborn para criar gráficos estatísticos elegantes
import matplotlib.pyplot as plt     # Importa o Matplotlib para configurar e exibir gráficos

# Seleciona as colunas categóricas (tipo 'object') do DataFrame
cat_cols = df.select_dtypes(include=['object']).columns

# Define uma paleta de cores suaves para os boxplots
palette = sns.color_palette("pastel")

# Loop para gerar um boxplot para cada coluna categórica
for i, col in enumerate(cat_cols):
    plt.figure(figsize=(7, 4))  # Define o tamanho da figura (largura x altura)
    
    # Cria um boxplot relacionando cada categoria da coluna com a variável 'monthly_sales'
    sns.boxplot(
        x=col,                     # Eixo X: variável categórica
        y='monthly_sales',         # Eixo Y: variável numérica (vendas mensais)
        data=df,                   # Fonte dos dados
        palette=palette,           # Cores suaves
        showmeans=True,            # Mostra o valor médio no gráfico
        meanprops={                # Personaliza o marcador da média
            "marker": "o",
            "markerfacecolor": "red",
            "markersize": 6
        }
    )
    
    # Define o título do gráfico com formatação de texto
    plt.title(
        f"Distribuição de Vendas por {col.replace('_', ' ').title()}",
        fontsize=12,
        fontweight='bold'
    )
    
    # Define o nome dos eixos X e Y com formatação amigável
    plt.xlabel(col.replace('_', ' ').title())
    plt.ylabel("Vendas Mensais")
    
    # Rotaciona os rótulos do eixo X para evitar sobreposição
    plt.xticks(rotation=30, ha='right')
    
    # Adiciona uma grade horizontal para facilitar a leitura dos valores
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    
    # Ajusta automaticamente os espaços e margens do gráfico
    plt.tight_layout()
    
    # Exibe o gráfico
    plt.show()
```
